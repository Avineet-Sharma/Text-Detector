# Using easyocr

```python
import easyocr
reader=easyocr.Reader(['en'],gpu=False)
results=reader.readtext('Task1\\1.jpg')
df=pd.DataFrame(results,columns=['bbox','text','confidence'])
print(df)
for text in results:
    print(text[1])
```

# Using keras_ocr

```python
import keras_ocr
pipeline=keras_ocr.pipeline.Pipeline()
img_fns=glob("Task1\\1.jpg")
results=pipeline.recognize(img_fns)
df=pd.DataFrame(results[0],columns=['text','bbox'])
print(df)
tc=df['text']
print(tc)
```
